```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.linear_model import LinearRegression


volcanoes = pd.read_csv("volcanoes.csv")

x = volcanoes["Distance (Km)"].to_numpy().reshape(-1, 1)
y = volcanoes["Age (Ma)"].to_numpy().reshape(-1, 1)

reg = LinearRegression().fit(x, y)

x_lbf = [min(x), max(x)]
y_lbf = reg.predict(x_lbf)

plt.scatter(x, y, color="black", marker="+")
plt.plot(x_lbf, y_lbf, color="red")

plt.title("Distance vs Age for Hawaiian Volcanoes")
plt.xlabel("Age (Ma)")
plt.ylabel("Distance (Km)")
plt.show()

print(reg.coef_)
```
